```python
import pathlib
import shutil
```

```python
pathlib.Path.cwd()
```

```python
for i in range(1,5):
    listfiles = list(pathlib.Path('.').glob(f'cam*_plane{i}*'))
    for file in listfiles:
        print(file)
        newfile = f'plane{i}_'+str(file).replace(f'_plane{i}','')
        print(f'copying {file}, {newfile}')
        shutil.copyfile(file, pathlib.Path(newfile))
    
```

```python
import numpy as np

# Read the original file
for k in range(1, 4):
    with open(f'../processed/c0/marker_c0_{k}.txt', 'r') as file:
        lines = file.readlines()

    # Remove the header
    lines = lines[1:]

    # Prepare the new data
    new_data = []
    for i, line in enumerate(lines):
        columns = np.array(line.split(), dtype=float)
        new_row = f"{i+1+k*1000:<10}{columns[2]:<10.1f}{columns[4]:<10.1f}{1400-columns[3]:<10.1f}"
        new_data.append(new_row)

    # Write the new data to a new file
    with open(f'plane_{k}.txt', 'w') as file:
        file.write('\n'.join(new_data))
```
